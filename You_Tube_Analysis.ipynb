**Emoji Analysis**

```python
!pip install emoji
import emoji
```

```python
comment  = "trending 😉"
[char for char in comment if char in emoji.EMOJI_DATA]
```

```python
emoji_list=[]
for comment in df['comment_text']:
  for char in comment:
      if char in emoji.EMOJI_DATA:
        emoji_list.append(char)
emoji_list
```

```python
from collections import Counter
Counter(emoji_list).most_common()
```

```python
emojis = []
frequencies = []

# Iterate through the input list and separate emojis and frequencies
for emoji in Counter(emoji_list).most_common():
    emojis.append(emoji[0])
    frequencies.append(emoji[1])

# Now, 'emojis' contains the emojis, and 'frequencies' contains their corresponding frequencies
print("Emojis:", emojis)
print("Frequencies:", frequencies)
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot
trace = go.Bar(x=emojis,y=frequencies)
iplot([trace])
```
